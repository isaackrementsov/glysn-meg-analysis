```python
import pandas as pd
import numpy as np
```

```python
merged_scores = np.load("./merged_scores.npy", allow_pickle=True)
```

```python
merged_scores
```

```python
merged_scores = merged_scores.item()
```

```python
for tone_type in ["pure", "complex", "ambiguous"]:
    tone_df = pd.DataFrame(merged_scores[tone_type])
    tone_df = tone_df.rename({col: f"T{col}" for col in tone_df.columns}, axis=1)
    tone_df.to_csv(f"{tone_type}_scores.csv", index=False)
```
